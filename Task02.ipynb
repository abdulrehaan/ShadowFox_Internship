```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
df['hour'] = df['date'].dt.hour
```

```python
print(df)
```

```python
pollutants=df[['co','no','no2','o3','so2','pm2_5','pm10','nh3']].sum()
```

```python
print(pollutants)
```

```python
plt.pie(pollutants, labels=pollutants.index, textprops={'rotation': 330})
plt.title('Pollutants Percentage')
plt.axis('equal')
```

```python

pollutants = ['co', 'no2', 'o3', 'so2', 'pm2_5', 'pm10', 'nh3']

for pollutant in pollutants:
    plt.figure(figsize=(10, 8))
    plt.hist(df[pollutant], bins=20, color='green', edgecolor='black')
    plt.xlabel(pollutant.upper())
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {pollutant.upper()}')
    plt.grid(True)
    plt.show()
```

```python
correlation_matrix = df.drop(columns='date').corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

df['season'] = df['month'].apply(get_season)
```

```python
# Summary statistics
print("Summary Statistics:\n", df.describe())
```

```python
# Plot 1: PM2.5 and PM10 over time
plt.figure(figsize=(14, 6))
plt.plot(df['date'], df['pm2_5'], label='PM2.5', alpha=0.7)
plt.plot(df['date'], df['pm10'], label='PM10', alpha=0.7)
plt.title('PM2.5 and PM10 Levels Over Time')
plt.xlabel('Date')
plt.ylabel('Concentration (µg/m³)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Seasonal variation in PM2.5
plt.figure(figsize=(8, 6))
sns.boxplot(x='season', y='pm2_5', data=df, legend=False, palette='Set2')
plt.title('Seasonal Variation in PM2.5 Levels')
plt.ylabel('PM2.5 (µg/m³)')
plt.xlabel('Season')
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Air Pollutants')
plt.tight_layout()
plt.show()
```
